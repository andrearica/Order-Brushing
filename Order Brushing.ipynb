```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('order_brush_order.csv',parse_dates=['event_time'])
df
```

```python
#88.2%
shopids = df['shopid'].unique()
final = pd.DataFrame(shopids,columns=['shopid'])
final['userid']=''
for i in range(len(shopids)):
    suspects = []
    df1 = df[(df['shopid']==shopids[i])]
    time = df1['event_time'].unique()
    for j in range(len(time)):
        interval = time[j] +pd.Timedelta(hours=1)
        df2 = df1[df1['event_time']<interval]
        df2 = df2[df2['event_time']>=time[j]]
        df3 = df2.pivot_table(index=['userid'], aggfunc='size').reset_index()
        ConcRate = df3[0].sum()/len(df3)
        if ConcRate>=3:
        #Order Proportion
            #print("conc")
            df3['OrderProp'] = df3[0]/df3[0].sum()
            suspects = (np.append(suspects,df3.loc[df3.OrderProp == df3.OrderProp.max(), 'userid'].values))

    if i%500==0:print(i)
    suspects = list(map(int,list(set(set(suspects)))))
    if len(suspects)==0:
        final.loc[final.shopid == shopids[i], "userid"] = 0
    elif len(suspects) == 1:
        final.loc[final.shopid == shopids[i], "userid"] = suspects
    else:
        list_suspects=suspects[0]
        for l in range(1,len(suspects)):
            list_suspects = "{}&{}".format(list_suspects,suspects[l])
        final.loc[final.shopid == shopids[i], "userid"] = list_suspects
```
